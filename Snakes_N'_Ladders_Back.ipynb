```python
# Librerías:

import numpy as np
import random
```

```python
# Definimos la clase jugador:

class Player:
  
  def __init__(self, Name):
        self.Name = Name  
        self.Position = 1
  
  # Añadimos el dado:
  def dice(self):
    return random.randint(1,6)    
```

##Simulación

```python
# Definimos el tablero:

class Board:
  def __init__(self, Num_Players, Board_Size):

    # Número de Jugadores:
    self.Num_Players = Num_Players
    self.Board_Size = Board_Size

    # Configuración de Serpientes y Escaleras:
    self.Snakes_100 = {22:2, 36:7, 42:40, 68:49, 77:28, 81:62, 86:74, 94:71, 96:84, 98:63}
    self.Ladders_100 = {6:16, 10:61, 18:38, 21:39, 30:50, 48:67, 57:65, 69:87, 76:95, 90:92}
  
  # # Función para definir quién va primero:
  # def first(self,p):

  #   trws = {}
  #   for i in range(len(p)):
  #     nombre = p[i].Name
  #     lanz = p[i].dice()
  #     print("{} sacó un {}".format(nombre, lanz) )
  #     trws[lanz] = nombre

  #   orden = [trws[i] for i in sorted(trws, reverse = True)]

  #   for i in range(len(orden)):
  #     p[i].Name = orden[i]

  #   return p

    
  def start_game(self):

    p = []
    for i in range(self.Num_Players):
      x = Player(input("Inserte su nombre "))
      if x.Name == '':
        x  = Player('Player_' + str(i+1))
      p.append(x)

    # p = self.first(p)
        
    while True:
      for i in p:
           print("Turno de {}".format(i.Name))
           
           tw = i.dice()
           print("Tu lanzamiento fue de {}".format(tw))

           if i.Position + tw > self.Board_Size:
             continue
           i.Position += tw 

           if i.Position in self.Snakes_100.keys():
                i.Position = self.Snakes_100[i.Position]
                
           if i.Position in self.Ladders_100.keys():
                i.Position = self.Ladders_100[i.Position]

           print("Tu posición es {}".format(i.Position))

           if i.Position == self.Board_Size:
            return "Ganador {}".format(i.Name)
```

```python
b = Board(3,100)
b.start_game()
```
